```python
import torch
import torch.utils.data
import torch.utils.data.distributed
import torchvision.transforms as transforms
import torchvision.datasets as datasets
import torchvision

from utils import show

image_data = torchvision.datasets.ImageFolder('dataset',
        transforms.Compose([transforms.ToTensor()])) # Récupération d'image en tenseurs

data_loader = torch.utils.data.DataLoader(image_data, batch_size=16,
                                            shuffle=True)
# Base de donné comportant le nombre d'image de batchsize
#Shuffle permet de prendre des images aléatoire a chaque époque

dataiter = iter(data_loader)
images,foo = next(dataiter)#Récupération de l'image et de l'index
images_crop = images[:,:,15:135,75:195]# Crop de l'image

show(torchvision.utils.make_grid(images_crop, normalize=True))
```
